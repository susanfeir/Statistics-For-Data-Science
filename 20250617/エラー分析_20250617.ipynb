# (2) そこは時差の関係で、同じ日付でも時間が異なるならば、一対一とならなく、結局欠損値として扱う。なので、下の三つのセルのコードを実行して、解決方法はcell 1 とcell 3二つ考えられる。

```python
# cell 1
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataT = yf.Ticker("7203.T").history(start='2019-01-01', end='2024-12-31')['Close']
dataG = yf.Ticker("GC=F").history(start='2019-01-01', end='2024-12-31')['Close']

# インデックスを日付のみに変換
dataT.index = dataT.index.date
dataG.index = dataG.index.date
print(dataT.index)

# pandasのデータフレーム用のDatetimeIndexに戻す
dataT.index = pd.to_datetime(dataT.index)
dataG.index = pd.to_datetime(dataG.index)

df1 = pd.DataFrame({'Toyota': dataT, 'Gold': dataG})
print(df1.head())
```
